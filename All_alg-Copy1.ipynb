```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df=pd.read_csv(r"Downloads\train.csv")
df
```

```python
sb.heatmap(df.corr(),annot=True)
```

```python
df.fillna(df.mean(),inplace=True)
df["Cabin"].fillna(df["Cabin"].mode()[0],inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
df.isnull().any()
```

```python
df=df.drop(labels=["PassengerId","Name","Ticket","Cabin"],axis=1)
df
```

```python
x=df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,6]=le.fit_transform(x[:,6])
x[:,1]=le.fit_transform(x[:,1])
x.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(categorical_features=[6])
x=ohe.fit_transform(x).toarray()
x.shape
```

```python
x[:,6]=x[:,7]
x[:,7]=x[:,8]
x=x[:,0:8]
x
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
from sklearn.metrics import accuracy_score
ac1=accuracy_score(y_test,y_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,p=2)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
from sklearn.metrics import accuracy_score
ac2=accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
s=SVC(kernel='linear')
s.fit(x_train,y_train)
y_pred=s.predict(x_test)
from sklearn.metrics import accuracy_score
ac3=accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
sv_rbf=SVC(kernel="rbf")
sv_rbf.fit(x_train,y_train)
y_predict_rbf=sv_rbf.predict(x_test)
ac4=accuracy_score(y_test,y_predict_rbf)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
from sklearn.metrics import accuracy_score
ac5=accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10)
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
from sklearn.metrics import accuracy_score
ac6=accuracy_score(y_test,y_pred)
```

```python
x=["LR","KNN","Linear SVM","Gaussian SVM","DT","RF"]
y=[ac1,ac2,ac3,ac4,ac5,ac6]
plt.bar(x,y)
plt.title("AUC",color='r')
```
